## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
unique_players = purchase_data.drop_duplicates(subset ="SN")
unique_players = unique_players.loc[:, ["Age"]]
total_players = unique_players["Age"].count()
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
label_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
unique_players["Age Group"] = pd.cut(unique_players["Age"], bins, labels=label_names)
unique_players
age_df = unique_players.groupby(["Age Group"])
age_demo_df = age_df.count()
age_demo_df.insert(1, "Percentage of Players", 0)
age_demo_df = age_demo_df.rename(columns={"Age": "Total Count"})
age_demo_df["Percentage of Players"] = age_demo_df["Total Count"] / total_players * 100 
age_demo_df["Percentage of Players"] = age_demo_df["Percentage of Players"].map("{0:.2f}%".format)
age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
label_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=label_names)
purchase_analysis = purchase_data.groupby(["Age Group"])
purchase_count = purchase_analysis.count()
purchase_count["Purchase ID"]
average_purchase_price = purchase_analysis["Price"].mean()
average_purchase_price
purchase_analysis.count()
merge_table_1 = pd.merge(purchase_count, average_purchase_price, on="Age Group")
total_purchase_value = purchase_analysis["Price"].sum()
merge_table_2 = pd.merge(merge_table_1, total_purchase_value, on="Age Group")
merge_table_2 = merge_table_2[["Price_x", "Price_y", "Price"]]
merge_table_2 = merge_table_2.rename(columns={"Price_x": "Purchase Count", "Price_y": "Average Purchase Price", "Price": "Total Purchase Value"})
merge_table_2.insert(3, "Avg Total Purchase per Person", merge_table_2["Total Purchase Value"] / age_demo_df["Total Count"])
merge_table_2["Average Purchase Price"] = merge_table_2["Average Purchase Price"].map("${:.2f}".format)
merge_table_2["Total Purchase Value"] = merge_table_2["Total Purchase Value"].map("${:.2f}".format)
merge_table_2["Avg Total Purchase per Person"] = merge_table_2["Avg Total Purchase per Person"].map("${:.2f}".format)
merge_table_2
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = purchase_data.groupby(["SN"])
total_purchase_value_top = top_spenders["Price"].sum()
purchase_count_top = top_spenders["Purchase ID"].count()
merged_table_uno = pd.merge(total_purchase_value_top, purchase_count_top, on="SN")
merged_table_uno = merged_table_uno.rename(columns={"Price": "Total Purchase Value", "Purchase ID": "Purchase Count"})
merged_table_uno.insert(1, "Average Purchase Price", merged_table_uno["Total Purchase Value"] / merged_table_uno["Purchase Count"])
merged_table_uno["Average Purchase Price"] = merged_table_uno["Average Purchase Price"].map("${:.2f}".format)
merged_table_uno = merged_table_uno[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
merged_table_uno = merged_table_uno.sort_values("Total Purchase Value", ascending=False)
merged_table_uno
```
